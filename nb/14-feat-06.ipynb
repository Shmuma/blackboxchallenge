```python
import sys
sys.path.append("..")
from lib import ipy
import matplotlib.pylab as pl
import numpy as np

%matplotlib inline
%load_ext autoreload
%autoreload 2
pl.rcParams['figure.figsize'] = [10.0, 8.0]
```

```python
feat_6 = ipy.load_feature(6)
ipy.overview(feat_6)
```

```python
mask_1 = ipy.bounds_feat(feat_6, (-1, 0))
sum(mask_1)
```

```python
ipy.scatter_mask(feat_6, mask_1)
```

```python
set_1 = set(feat_6[1][mask_1])
len(set_1), set_1
```

```python
full_set = set(feat_6[1])
len(full_set)
```

```python
vc_full = feat_6[1].value_counts()
centers = ipy.cluster_1D(vc_full.index)
deltas = ipy.deltas(centers)
```

## Stripe 1

```python
start = 1
stop = start + 59
deltas_ = deltas[start:stop]
print "len=%d, min=%f, max=%f" % (len(deltas_), min(deltas_), max(deltas_))
print "prev=%f, succ=%f" % (deltas[start-1], deltas[stop])
np.mean(deltas_)
```

```python
centers_ = centers[start:stop+1]
print "len=%d, min=%f, max=%f, prev=%f, succ=%f" % (len(centers_), min(centers_),
                                                    max(centers_), centers[start-1], centers[stop+1])
min(centers_), max(centers_)
```
