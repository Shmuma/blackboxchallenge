```python
import sys
sys.path.append("..")
from lib import ipy
import matplotlib.pylab as pl
import numpy as np

%matplotlib inline
%load_ext autoreload
%autoreload 2
pl.rcParams['figure.figsize'] = [10.0, 8.0]
```

```python
feat_5 = ipy.load_feature(5)
ipy.overview(feat_5)
```

```python
mask_1 = ipy.bounds_feat(feat_5, (-1, 0))
sum(mask_1)
```

```python
ipy.scatter_mask(feat_5, mask_1)
```

```python
set_1 = set(feat_5[1][mask_1])
len(set_1), set_1
```

```python
full_set = set(feat_5[1])
len(full_set)
```

```python
vc_full = feat_5[1].value_counts()
```

```python
centers = ipy.cluster_1D(vc_full.index)
```

```python
deltas = ipy.deltas(centers)
```

## Stripe 1

```python
np.mean(deltas[1:60])
```

```python
centers_1 = centers[1:61]
min(centers_1), max(centers_1)
```

```python
m = (feat_5[1] >= min(centers_1)) & (feat_5[1] <= max(centers_1))
count = sum(m)
print "%d (%.4f%%)" % (count, 100.0 * count / len(feat_5))
```
