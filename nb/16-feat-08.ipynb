```python
import sys
sys.path.append("..")
from lib import ipy
import matplotlib.pylab as pl
import numpy as np

%matplotlib inline
%load_ext autoreload
%autoreload 2
pl.rcParams['figure.figsize'] = [10.0, 8.0]
```

```python
feat = ipy.load_feature(8)
ipy.overview(feat)
```

```python
mask = ipy.bounds_feat(feat, (-1, -0.5))
sum(mask)
```

```python
ipy.scatter_mask(feat, mask)
```

```python
set(feat[1][mask])
```

```python
centers = ipy.cluster_1D(feat[1])
deltas = ipy.deltas(centers)
```

## Stripe 1

```python
start = 1
stop = start + 59
deltas_ = deltas[start:stop]
print "len=%d, min=%f, max=%f" % (len(deltas_), min(deltas_), max(deltas_))
print "prev=%f, succ=%f" % (deltas[start-1], deltas[stop])
np.mean(deltas_)
```

```python
centers_ = centers[start:stop+1]
print "len=%d, min=%f, max=%f, prev=%f, succ=%f" % (len(centers_), min(centers_),
                                                    max(centers_), centers[start-1], centers[stop+1])
min(centers_), max(centers_)
```

```python
m = (feat[1] >= min(centers_)) & (feat[1] <= max(centers_))
count = sum(m)
print "%d (%.4f%%)" % (count, 100.0 * count / len(feat))
```
